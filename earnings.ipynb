# Lifetime Earnings: U of L instructors

This notebook will explore the lifetime earnings for an Instructor at the University of Lethbridge.

There are three Instructor ranks at U of L: I, II, and III. However, these ranks are not based on career progress, like those for Professors, who begin at Assistant, and can then be promoted to Associate, and eventually, Full Professor.

Instead, the Instructor ranks reflect qualifications and duties. Recently, there have been no new hires at the rank of Instructor I, so we will omit them from the comparison. In Arts and Sciences, an Instructor II is typically someone with a Master's degree, or equivalent experience, whose duties consist primarily of teaching labs or tutorials. An Instructor III is usually someone with a PhD, whose duties consist primilary of teaching courses.

In Instructor is not expected to engage in research activities (unlike professors), but many Instructors engage in Scholarship of Teaching and Learning, curriculum development, textbook writing, and other scholarly activity that is often not officially considered part of their workload.

Professors at U of L see their salaries increase through a combination of career progress and merit. The career progress increase reflects the increased experience of the professor, and progress through the ranks. Merit is strictly performance based. Since Instructors usually remain at the same rank for their career, they do not receive career progress. Instead, their increase is based entirely on merit.

The merit amount for professors is $\$2800$, or $\$1400$ per merit unit. (Merit scores range from 0.0 to 2.0 in increments of 0.25.) This is paid as a contribution to a **merit pool**. However, unlike professors, Instructors have a salary cap. Under the current collective agreement, salary ranges are:

- Instructor II: $\$56,500 -- \$119,443$
- Instructor III: $\$61,500 -- \$126,588$

Instructors at the cap receive merit pay as a lump sum payment, but do not see an increase in salary. Their contributions are therefore removed from the merit pool. However, when the value of a merit unit is calculated, the university divides the total value of the merit pool by the sum of all merit amounts awarded, including those at the cap. This means that for an Instructor earning the maximum merit score of 2.0, the value of the merit increase is closer to $\$2500$ than to $\$2800$. The exact amount varies from year to year, but the average over the last 5 years is close to $\$2500$, so we will use that as our amount.

We will consider three scenarios, for both an Instructor II and an Instructor III:

1. An Instructor with "acceptable" performance, with an average merit score of 1.0 (merit amount $\$1250$)
2. An Instructor with "good" performance, with an aveverage merit score of 1.5 (merit amount $\$1875$)
3. An Instructor with "excellent" performance, with an average merit score of 2.0 (merit amount $\$2500$)

## Instructor II

For both Instructor II and Instructor III, we will assume the following for undergraduate studies:

- Tuition of $\$6000$ per year
- Part-time income of $\$10000$ per year covers rent, and some food
- Total student loan of $\$30000$
- Other debt of $\$10000$ due to costs for food and entertainment

For the Instructor II, we assume a two-year Master's degree, unfunded, at $\$10000$ tuition. (Many graduate programs do not provide funding for students not proceeding to PhD, especially in Arts and Humanities.) Note that undergraduate student loans do not need to be paid while in graduate school.

- Tuition of $\$10000$ per year
- Part-time income of $\$15000$
- Non-student loan debt servicing of $\$2500$ per year
- Student loan of $\$25000$, resulting in $\$55000$ total
- Additional debt of $\$5000$ incurred

According to the [Student loan repayment estimator](https://tools.canlearn.ca/cslgs-scpse/cln-cln/crp-lrc/af.nlindex-eng.do) provided by the Government of Canada, our Instructor II can expect debt servicing costs of about $\$500$ per month, or $\$6000$ per year, for 15 years, assuming an interest rate of 3%. Servicing of additional debt will cost about $\$2500$ per year for an additional 5 years, assuming an interest rate of 8%.

Using data compiled by ULFA, the average starting age for an Instructor II at U of L is 45 years old. This leaves us with a gap of about 20 years to account for between the end of their Master's degree and the beginning of their employment as an Instructor II! However, this average is skewed somewhat by Instructors who have taken a job in one of the professional faculties (Education, Health Sciences, Business) after a career in industry.

We will instead assume an Instructor in Arts and Sciences. It is unlikely that they are lucky enough to land their job directly out of their Master's degree. Most will have 5-10 years of employment in related (or unrelated) jobs, and possibly some term or sessional appointments. These years are the most difficult to estimate, since they vary widely from person to person. Median employment income in Lethbridge is [reported by Statistics Canada](https://www12.statcan.gc.ca/census-recensement/2016/dp-pd/prof/details/page.cfm?Lang=E&Geo1=CMACA&Code1=810&Geo2=PR&Code2=47&SearchText=Nobleford&SearchType=Begins&SearchPR=01&B1=Income&TABID=1&type=0) to be about $\$54,500$.

Our early career individual probably makes below median. Let's take an average income of $\$45,000$. For the first 5 years, they will be paying debt servicing of $\$8500$, leaving a net income of $\$36,500$. For the next 5 years, debt servicing drops to $\$6000$.

Let's say our Instructor II begins at age 34, with a starting salary of $\$60,000$. They still have 5 years of student loan payments at this point. The following data assumes employment to age 65, with the three merit categories given above. Note that only the Instructor II earning the maximum merit amount reaches the salary cap. Once they reach the cap, they receive their merit pay as a lump sum amount, rather than a salary increase.

## Instructor III

For an Instructor III, we will assume 6 years of graduate school (2 Master's and 4 PhD). This time, we will assume that our Instructor III received funding in graduate school, although this is not always the case.

A typical stipend for a graduate student is around $\$10000$ after tuition. Our graduate student will have annual debt servicing costs of $\$2500$ for non-student loan debt, and will accrue an additional $\$5000$ per year in new debt. 

At the end of their PhD, they will owe $\$30,000$ in student loans, and about the same amount in other debt, because the stipend income they receive is not quite enough to cover cost of living. Assuming repayment over 10 years, debt servicing will be about $\$600$ per month, or $\$7200$ anually.

Post PhD, our Instructor III will land two postdoctoral positions, each two years in duration. The typical postdoctoral salary in Canada is about $\$40,000$. After that, we include one year as a sessional lecturer (income $\$20,000$, and two years on term teaching contracts (income $\$50,000$).

After 4 years of undergraduate education, 6 years of graduate education, 4 years postdoctoral studies, and 3 years in teaching-related (but temporary) positions, our Instructor III begins at U of L at age 35, with a starting salary of $\$65,000$. 

Again, only the instructor who receives the maximum merit award each year reaches the salary cap, which they do at age 60.

```python
I3 = pd.read_csv("InstructorIII.csv",names=['Age','Merit 1.0','Merit 1.5','Merit 2.0'])
I3
```

```python
I3[['Merit 1.0','Merit 1.5','Merit 2.0']].sum()
```
